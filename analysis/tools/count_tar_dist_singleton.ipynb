```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
```

```python
angles_task_gabors = [
            i
            for i in np.linspace(
                0,
                360,
                8,
                endpoint=False,
            ) % 360
        ]

HPL = 0

# the first column is the angle of the target, the second column is the angle of the distractor
TD_pattern = list(itertools.permutations(angles_task_gabors, 2))
TD_pattern_HPL = np.array(
    [(angle, HPL) for angle in angles_task_gabors if angle != HPL]
)
TD_pattern_HPL = np.tile(
    TD_pattern_HPL, (len(angles_task_gabors) - 1, 1)
)
TD_pattern_HPL_additional = np.array(
    [(HPL, angle) for angle in angles_task_gabors if angle != HPL]
)
TD_pattern_HPL = np.vstack(
    (TD_pattern_HPL, TD_pattern_HPL_additional)
)

if HPL is None:
    TD_pattern = np.tile(TD_pattern, (2, 1))
elif HPL is not None:
    TD_pattern = np.vstack((TD_pattern, TD_pattern_HPL))
```

```python
plt.hist(TD_pattern[:, 0], bins=15, alpha=0.5)
plt.title("Target")
plt.show()
```

```python
plt.hist(TD_pattern[:, 1], bins=15, alpha=0.5)
plt.title("Distractor")
plt.show()
```
