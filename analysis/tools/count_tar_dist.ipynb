```python
import numpy as np
import itertools
from copy import deepcopy
import matplotlib.pyplot as plt
```

```python
angles_gabors = [
            int(i) for i in np.linspace(45, 360 + 45, 4, endpoint=False)
        ] % np.array([360])
TD_pattern = list(itertools.permutations(angles_gabors, 2))
TD_pattern = np.tile(TD_pattern, (4, 1))
HPL = angles_gabors[0]
for i in range(int(len(TD_pattern) / 2)):
    if TD_pattern[i, 0] == HPL:
        tmp = deepcopy(TD_pattern[i, 0])
        TD_pattern[i, 0] = TD_pattern[i, 1]
        TD_pattern[i, 1] = tmp
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
# plot histogram
ax.hist(TD_pattern[:,1], bins=20, histtype="bar")
ax.set_yticks(np.arange(0, 20, 1))
ax.set_ylabel("Freq of distractor")
plt.show()
```

```python
# check frequency of distractor of each angle
mydict = {}
unique, counts = np.unique(TD_pattern[:,1], return_counts=True)
mydict = dict(zip(unique, counts))
```

```python
print(f"Freq of HPL_D: {np.max(list(mydict.values())) / np.sum(list(mydict.values())) * 100:.2f}")
```

```python
print(f"Freq of LPL_D: {np.min(list(mydict.values())) / np.sum(list(mydict.values())) * 100:.2f}")
```
